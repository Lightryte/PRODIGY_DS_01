```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import argparse as ap
```

```python
data = pd.read_csv('/content/exercise.csv')
```

```python
data.info()
```

```python
plt.figure(figsize=(10,1))
sns.countplot(data['Gender'], palette='Set3')
plt.title('Gender Distibution of Exercise')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10,2))
sns.histplot(data['Age'], bins=8, kde=False, color='skyblue')
plt.title('Age Distribution of Exercise')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```
